```python
# SVE U JEDNOJ ĆELIJI - OD POČETKA DO KRAJA

import pandas as pd
import requests
from sqlalchemy import create_engine
from datetime import date

# --- EXTRACT ---
print("1. Extract: Povlačim podatke sa API-ja...")
url = 'https://jsonplaceholder.typicode.com/users'
response = requests.get(url)
users_df = pd.DataFrame(response.json())

# --- TRANSFORM ---
print("2. Transform: Sređujem podatke...")
potrebne_kolone = ['id', 'name', 'username', 'email', 'address']
cleaned_users_df = users_df[potrebne_kolone].copy()
cleaned_users_df['city'] = cleaned_users_df['address'].apply(lambda x: x['city'])

# Ključni korak koji smo dodali:
cleaned_users_df = cleaned_users_df.drop('address', axis=1) 

cleaned_users_df['import_date'] = date.today()
print("   - Podaci sređeni. Finalni izgled:")
print(cleaned_users_df.head())

# --- LOAD ---
print("3. Load: Upisujem podatke u bazu...")
# !!! ZAMENI 'TvojaLozinka!' SA SVOJOM LOZINKOM !!!
connection_string = 'postgresql://postgres:Milos098!@localhost:5432/leetcode_practice'
engine = create_engine(connection_string)

cleaned_users_df.to_sql(
    'api_users',
    engine,
    if_exists='replace',
    index=False
)

print("\nUSPEH! Proces završen. Proveri DBeaver sada (nakon što uradiš Refresh).")
```
