```python
from pyspark.sql import SparkSession

# Korak 1: Kreiranje Spark Sesije
# Ovo je ulazna tačka u Spark svet, kao 'engine' za bazu
# Kreira mali, lokalni Spark "klaster" unutar tvog računara
spark = SparkSession.builder.appName("UvodUSpark").getOrCreate()

print("Spark sesija uspešno napravljena!")
print(spark)

# Korak 2: Učitavanje ISTOG CSV fajla, ali sada sa Spark-om
# Zameni ime fajla ako je potrebno
spark_df = spark.read.csv(
    "IMDB Movies 2000 - 2020.csv",
    header=True,       # Kažemo Sparku da prvi red sadrži imena kolona
    inferSchema=True   # Kažemo Sparku da pokuša sam da pogodi tipove podataka (npr. da je 'year' broj)
)

# Korak 3: Prikazivanje podataka
print("\nShema (struktura) Spark DataFrame-a:")
spark_df.printSchema() # Slično kao .info() u Pandas-u

print("\nPrvih 5 redova podataka:")
spark_df.show(5) # Slično kao .head() u Pandas-u

# Korak 4: Zaustavljanje sesije (dobra praksa)
spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Kreiramo Spark Sesiju
spark = SparkSession.builder.appName("Transformacije").getOrCreate()

# Učitavamo podatke
spark_df = spark.read.csv(
    "IMDB Movies 2000 - 2020.csv",
    header=True,
    inferSchema=True
)

# --- TRANSFORMACIJE ---

# 1. Selekcija kolona (slično kao df[['col1', 'col2']])
# Ovo je samo plan, ništa se još ne izvršava
selected_df = spark_df.select("title", "year", "avg_vote")

# 2. Filtriranje (slično kao df[df['year'] > 2015])
# I ovo je samo plan, dodajemo novi korak
filtered_df = selected_df.filter(col("year") > 2015)

# 3. Kreiranje nove kolone (slično kao df['new_col'] = ...)
# I dalje samo planiramo
df_with_new_col = filtered_df.withColumn("ocena_puta_10", col("avg_vote") * 10)

# 4. Prikazivanje rezultata - AKCIJA!
print("Rezultat nakon svih transformacija:")
# TEK SADA, kada pozovemo .show(), Spark pogleda ceo plan i izvrši sve odjednom
df_with_new_col.show(10)

# Zaustavljamo sesiju
spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.sql.types import DoubleType # Važan novi import za definisanje tipa

# Kreiramo Spark Sesiju
spark = SparkSession.builder.appName("Transformacije").getOrCreate()

# Učitavamo podatke
spark_df = spark.read.csv(
    "IMDB Movies 2000 - 2020.csv",
    header=True,
    inferSchema=True
)

# --- PRIPREMA TIPOVA PODATAKA ---
print("\nPretvaram 'avg_vote' kolonu u numerički tip...")
# Kažemo Sparku da prepiše kolonu 'avg_vote' sa novom verzijom
# gde je svaka vrednost pretvorena (cast) u Double (decimalni broj)
spark_df = spark_df.withColumn("avg_vote", col("avg_vote").cast(DoubleType()))


# --- TRANSFORMACIJE ---
# Ostatak koda ostaje POTPUNO ISTI

# 1. Selekcija kolona
selected_df = spark_df.select("title", "year", "avg_vote")

# 2. Filtriranje
filtered_df = selected_df.filter(col("year") > 2015)

# 3. Kreiranje nove kolone
df_with_new_col = filtered_df.withColumn("ocena_puta_10", col("avg_vote") * 10)

# 4. Prikazivanje rezultata - AKCIJA!
print("\nRezultat nakon svih transformacija:")
df_with_new_col.show(10)

# Zaustavljamo sesiju
spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Kreiramo Spark Sesiju
spark = SparkSession.builder.appName("Agregacije").getOrCreate()

# Učitavamo podatke
spark_df = spark.read.csv(
    "IMDB Movies 2000 - 2020.csv",
    header=True,
    inferSchema=True
)

# --- AGREgACIJA ---
print("\nBroj filmova po godini:")

# Slično kao u Pandas-u: .groupby() pa agregatna funkcija .count()
movies_per_year_df = spark_df.groupBy("year").count()

# Hajde da sortiramo rezultat da bude preglednije (od najnovijih godina)
sorted_movies_per_year = movies_per_year_df.orderBy(col("year").desc())

# AKCIJA! Prikazujemo rezultat
sorted_movies_per_year.show(10)

# Zaustavljamo sesiju
spark.stop()
```

```python
from pyspark.sql import SparkSession

# Kreiramo Spark Sesiju
spark = SparkSession.builder.appName("SparkSQL").getOrCreate()

# Učitavamo podatke
spark_df = spark.read.csv(
    "IMDB Movies 2000 - 2020.csv",
    header=True,
    inferSchema=True
)

# --- SPARK SQL ---

# Korak 1: Registruj DataFrame kao privremenu tabelu
# Dajemo joj ime 'movies_table' koje ćemo koristiti u SQL-u
spark_df.createOrReplaceTempView("movies_table")

print("DataFrame 'spark_df' je sada dostupan kao SQL tabela 'movies_table'.")

# Korak 2: Napiši SQL upit kao običan Python string
# Ovo je standardni SQL koji si već naučio!
sql_upit = """
    SELECT
        year,
        COUNT(*) AS broj_filmova
    FROM
        movies_table
    WHERE
        year IS NOT NULL
    GROUP BY
        year
    ORDER BY
        year DESC
"""

# Korak 3: Izvrši SQL upit pomoću spark.sql()
# Rezultat je NOVI Spark DataFrame
rezultat_df = spark.sql(sql_upit)

# Korak 4: Prikaži rezultat
print("\nRezultat dobijen pomoću Spark SQL-a:")
rezultat_df.show(10)

# Zaustavljamo sesiju
spark.stop()
```
